```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold
```

```python
data = pd.read_csv('/content/drive/MyDrive/final_data.csv', header = 0)
```

```python
X = data.iloc[:, 1:]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.metrics import silhouette_score
```

```python
 def clustering(X, cluster_type, scaling_type):

      if scaling_type == 'MinMax':
        scaler = MinMaxScaler()
      elif scaling_type == 'Standard':
        scaler = StandardScaler()
      X = scaler.fit_transform(X)

      n_clusters = [2,3,4,5,6]
      score_list = []
      for n  in n_clusters:

          if cluster_type == 'Kmeans':
              cluster = KMeans(n_clusters=n, init='kmeans++', random_state=2023)
          elif cluster_type == 'Hierarchical':
              cluster = AgglomerativeClustering(n_clusters=n, linkage='complete')
          else:
              print('Check Model Type')
              raise NameError

          cluster.fit(X)
          c_label = cluster.labels_
          score = silhouette_score(X, c_label)
          score_list.append(score)

        best_n = n_clusters[np.argmax(score_list)] #4
        print(f'Optimal number of cluster - {cluster_type}: {best_n}')

        plt.figure(figsize=(8,4))
        plt.plot(score_list, marker='o', color = 'gray', lw=1, ls = '--')
        plt.title(f'Silhouette Score Plot:{cluster_type}')
        plt.xlabel('n_clusters')
        plt.xticks(np.arange(len(n_clusters)),n_clusters)
        plt.show()
```
