```python
import pandas as pd
from sklearn.cluster import KMeans

data = pd.read_csv("C:/Users/dong_/OneDrive/바탕 화면/K-디지털 프로젝트/final_data.csv")

# k-means 클러스터링 객체 생성 및 학습
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(data)

# 클러스터 할당 결과 확인
labels = kmeans.labels_
print(labels)

# 클러스터 중심 확인
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels)  # .iloc를 사용하여 인덱싱
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='red', s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.show()
```
