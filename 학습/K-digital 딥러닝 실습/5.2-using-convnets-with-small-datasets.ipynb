## The relevance of deep learning for small-data problems

딥 러닝은 많은 데이터를 사용할 수있을 때만 작동한다는 말을 듣게 될 것입니다. 이것은 부분적으로 유효합니다. 딥 러닝의 기본 특성 중 하나는 자체적으로 학습 데이터에서 흥미로운 패턴들을 찾을 수 있다는 것입니다. 이는 일반적으로 많은 학습 데이터가 있을 때 가능합니다. 입력 샘플이 이미지와 같이 매우 높은 차원 인 문제에서 특히 이러한 현상을 관측할 수 있습니다.

그러나 "많은" 샘플을 구성하는 것은 훈련하려는 네트워크의 크기와 깊이에 상대적입니다. 수십 개의 샘플만으로 복잡한 문제를 해결하도록 convnet을 훈련시키는 것은 불가능하지만 모델이 작고 잘 정규화되어 있고 작업이 간단한 경우에는 수백 개만으로 충분할 수 있습니다. convnet은 지역적이고 위치에 상관없는 패턴들을 학습하기 때문에 이미지 문제에 대해 매우 효율적입니다. 심지어 매우 작은 이미지 데이터 세트에서 convnet을 처음부터 학습하면  상대적인 데이터 부족에도 불구하고 비교적 합리적인 결과를 얻을 수 있습니다.

더군다나 딥 러닝 모델은 본질적으로 매우 다양한 용도로 변경 될 수 있습니다. 예를 들어 대규모 데이터 세트에서 훈련 된 이미지 분류 또는 음성-텍스트 모델을 취한 다음 fine tuning만으로 상당히 다른 문제에 재사용 할 수 있습니다. 특히 컴퓨터 비전의 경우 많은 사전 훈련 된 모델 (일반적으로 ImageNet 데이터 세트에서 훈련 됨)이 이제 공개적으로 다운로드 가능하며 매우 적은 데이터로 강력한 비전 모델을 부트 스트랩하는 데 사용할 수 있습니다. 이것은 다음 섹션에서 다뤄질 예정입니다.

## Building our network

이전 예에서 MNIST 용 소규모 convnet을 이미 구축 했으므로 전체적인 흐름은 익숙하실 겁니다. 동일한 일반 구조를 재사용합니다. convnet은 Conv2D (relu 활성화 포함) 및 MaxPooling2D 레이어의 스택이됩니다.

그러나 우리는 더 큰 이미지와 더 복잡한 문제를 다루고 있기 때문에 그에 따라 네트워크를 더 크게 만들 것입니다. Conv2D + MaxPooling2D 단계가 하나 더있을 것입니다. 이는 네트워크의 용량을 늘리고 feature map의 크기를 더 줄여서 Flatten 레이어에 도달했을 때 입력변수 크기가 지나치게 크지 않도록합니다. 여기서는 150x150 (다소 임의의 선택) 크기의 입력에서 시작하므로 Flatten 레이어 바로 전에 크기 7x7의 피처 맵으로 끝납니다.

feature map의 깊이는 네트워크에서 점진적으로 증가하는 반면 (32에서 128으로) feature map의 크기는 148x148에서 7x7로 감소합니다. 이것은 거의 모든 convnet에서 볼 수있는 패턴입니다.

이진 분류 문제를 해결하고 있기 때문에 single unit 'Dense' layer와
'sigmoid' 비선형 함수를 사용합니다. 이 단위는 네트워크가 한 클래스 또는 다른 클래스를 선택할 확률을 계산합니다.

```python
from tensorflow.keras import layers
from tensorflow.keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

Let's take a look at how the dimensions of the feature maps change with every successive layer:

```python
model.summary()
```

컴파일 단계에서는 평소와 같이`RMSprop` 옵티 마이저를 사용합니다. 단일 시그모이드 함수로 네트워크를 종료 했으므로
이진 크로스 엔트로피를 손실 함수로 사용하십시오.

```python
from tensorflow.keras import optimizers

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(learning_rate=1e-4),
              metrics=['acc'])
```

## Data preprocessing

이미 알고 있듯이 데이터는 적절하게 사전 처리 된 부동 소수점 텐서로 포맷팅 되어야합니다. 현재 데이터는 JPEG 파일로 드라이브에 저장되므로 네트워크로 가져 오는 단계는 대략 다음과 같습니다:

* 사진 파일 읽기
* JPEG 파일을 RGB 그리드로 변환하기
* 이를 floating point tensor로 변환하기
* 픽셀값들을 [0, 1] 사이 값으로 스케일 조정하기.

다소 힘들어 보일 수 있지만 다행히 Keras에는 이러한 단계를 자동으로 처리하는 유틸리티가 있습니다. Keras에는 이미지 처리 모듈이 있습니다.
`keras.preprocessing.image`에있는 처리 도우미 도구입니다. 특히 'ImageDataGenerator'클래스를 포함하여
디스크의 이미지 파일을 사전 처리 된 텐서의 배치 파일로 자동 변환 할 수있는 생성기를 빠르게 설정합니다. 

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='binary')
```

데이터 생성기 중 하나의 출력을 살펴 보겠습니다. 150x150 RGB 이미지 (모양`(20, 150, 150, 3)`) 및 바이너리 배치를 생성합니다.
레이블 (`(20,)`모양) 20은 각 배치의 샘플 수 (배치 크기)입니다. 생성기는 이러한 배치를 계속해서 생성합니다:대상 폴더에있는 이미지를 끝없이 반복합니다. 이러한 이유로 어느 시점에서 반복 루프를 '중단'해야합니다.

```python
for data_batch, labels_batch in train_generator:
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    break
```

생성기를 사용하여 모델을 학습시킵니다. 데이터 생성기의 'fit'과 동일한 'fit_generator'메서드를 사용하여 수행합니다. 데이터가 끝없이 생성되기 때문에 생성기는 이전에 생성기에서 몇 개의 샘플을 추출해야하는지 알아야합니다.이것이`steps_per_epoch` 인수의 역할입니다. 다음에서`steps_per_epoch` 배치를 가져온 후
즉, 'steps_per_epoch'개수 만큼 경사 하강 법 단계를 실행합니다. 우리의 경우 배치는 20 개이므로 2,000 개 샘플을 모두 학습시키기 위해 100 개 배치가 필요합니다.

`fit_generator`를 사용할 때 `fit` 메소드와 마찬가지로 `validation_data` 인수를 전달할 수 있습니다. 중요한 것은 데이터 생성기 자체가 될 수 있지만 Numpy 배열의 튜플이 될 수도 있습니다. 생성기를`validation_data`로 전달하면 이 생성기는 일련의 검증 데이터를 끝없이 생성 할 것으로 예상되므로 `validation_steps` 인수도 지정해야합니다.
평가를 위해 유효성 검사 생성기에서 가져올 배치 수를 프로세스에 알려줍니다.

```python
history = model.fit(
      train_generator,
      steps_per_epoch=100,
      epochs=20,
      validation_data=validation_generator,
      validation_steps=50)
```

## Using data augmentation

과적 합은 학습 할 샘플이 너무 적기 때문에 새 데이터로 일반화 할 수 있는 모델을 훈련 할 수 없게 됩니다. 무한한 데이터가 주어지면 우리 모델은 현재 데이터 분포의 가능한 모든 측면에 노출 될 것입니다. 우리는 결코 과적 합되지 않을 것입니다. Data Augmentation은 기존 훈련 샘플에서 더 많은 훈련 데이터를 생성하는 접근 방식을 취합니다.

`ImageDataGenerator`를 활용해서 다음과 같이 Data Augmentation을 할 수 있습니다:

```python
datagen = ImageDataGenerator(
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')
```

Tensorflow Keras ImageGenerator는 다음 인수들을 입력합니다:

* `rotation_range` is a value in degrees (0-180), a range within which to randomly rotate pictures.
* `width_shift` and `height_shift` are ranges (as a fraction of total width or height) within which to randomly translate pictures 
vertically or horizontally.
* `shear_range` is for randomly applying shearing transformations.
* `zoom_range` is for randomly zooming inside pictures.
* `horizontal_flip` is for randomly flipping half of the images horizontally -- relevant when there are no assumptions of horizontal 
asymmetry (e.g. real-world pictures).
* `fill_mode` is the strategy used for filling in newly created pixels, which can appear after a rotation or a width/height shift.

Augmented 된 이미지들은 다음과 같습니다:

```python
# This is module with image preprocessing utilities
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

fnames = [os.path.join(train_cats_dir, fname) for fname in os.listdir(train_cats_dir)]

# We pick one image to "augment"
img_path = fnames[3]

# Read the image and resize it
img = image.load_img(img_path, target_size=(150, 150))

# Convert it to a Numpy array with shape (150, 150, 3)
x = image.img_to_array(img)

# Reshape it to (1, 150, 150, 3)
x = x.reshape((1,) + x.shape)

# The .flow() command below generates batches of randomly transformed images.
# It will loop indefinitely, so we need to `break` the loop at some point!
i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break

plt.show()
```

적은 원본 이미지에서 비롯된 것이기 때문에 여전히 상호 연관성이 높은 것으로 보입니다. 새로운 정보를 생성 할 수 없습니다.
기존 정보만 재생성 할 수 있습니다. 따라서 이것은 과적 합을 완전히 제거하기에 충분하지 않을 수 있습니다. 과적합을 더 줄이기 위해
Densely connected layer 바로 앞에 Dropout 레이어를 모델에 추가합니다:

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(learning_rate=1e-4),
              metrics=['acc'])
```

Let's train our network using data augmentation and dropout:

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,)

# Note that the validation data should not be augmented!
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='binary')

history = model.fit(
      train_generator,
      steps_per_epoch=100,
      epochs=50,
      validation_data=validation_generator,
      validation_steps=50)
```
