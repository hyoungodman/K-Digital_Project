```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score

from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
```

```python
data = pd.read_csv('../data/final_data.csv')
```

```python
X = data.iloc[:, 1:]
y = data['TARGET']
y = y.replace(1,-1)
y = y.replace(0,1)

train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.3, stratify = y, random_state = 3000 )
cv = StratifiedKFold(n_splits = 3)
```

```python
def one_class_classification(model_name, train_x, train_y, cv):

    if model_name == 'iforest':
        model = IsolationForest()
        param_space = {'contamination': [0.1, 0.3, 0.5],
                       'n_estimators': [50, 75, 100],
                       'max_features': [0.3, 0.5, 0.8, 1]}

    grid_search = GridSearchCV(model, param_grid = param_space, cv = cv,
                                scoring = ['f1', 'precision', 'recall', 'accuracy'], refit='recall', n_jobs = -1)
    grid_search.fit(train_x, train_y)

    best_model = grid_search.best_estimator_

    return best_model
```

```python
def model_evaluation(true, pred):

  print('Confusion Matrix')
  print(confusion_matrix(true, pred, labels=[1,-1]))
  print('\n')
  print(f'Accuracy: {accuracy_score(true, pred)}')
  print(f'Recall: {recall_score(true, pred, pos_label=-1)}') # Pred-이상 / True-Total-이상 (실제로 정상 관측치 중에서 모델이 정상으로 예측한 비율)
  print(f'Precision: {precision_score(true, pred, pos_label=-1)}') # True-이상 / Pred-Total-이상 (모델이 정상으로 판단한 관측치 중 실제로 정상치였던 비율)
  print(f'F1:{f1_score(true, pred, pos_label=1)}') # Precision과 Recall의 조화평균
```

```python
def evaluate_model(y_true, y_pred, model_name):
    print(f"Performance Metrics for {model_name}:")
    print(classification_report(y_true, y_pred))
```

```python
iforest_model = one_class_classification('iforest', train_x, train_y, cv)
```

```python
pred_train_y_iforest = iforest_model.predict(train_x)
pred_test_y_iforest = iforest_model.predict(test_x)
```

```python
model_evaluation(test_y, pred_test_y_iforest)
```

```python
model_evaluation(train_y, pred_train_y_iforest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

evaluate_model(test_y, pred_test_y_iforest, iforest_model)
```

```python
evaluate_model(train_y, pred_train_y_iforest, iforest_model)
```

```python
def plot_confusion_matrix(true, pred):
    cm = confusion_matrix(true, pred, labels=[1, -1])
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted 1', 'Predicted -1'], yticklabels=['Actual 1', 'Actual -1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()


plot_confusion_matrix(test_y, pred_test_y_iforest)
```

```python
    from sklearn.metrics import roc_curve, auc, precision_recall_curve
    import matplotlib.pyplot as plt

    def plot_roc_curve(true, pred_probabilities):
        fpr, tpr, _ = roc_curve(true, pred_probabilities, pos_label=-1)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc='lower right')
        plt.show()

    def plot_precision_recall_curve(true, pred_probabilities):
        precision, recall, _ = precision_recall_curve(true, pred_probabilities, pos_label=-1)

        plt.figure(figsize=(8, 6))
        plt.plot(recall, precision, color='darkorange', lw=2)
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curve')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.show()

    # 모델의 이상치 확률 값을 얻습니다. (best_model은 one_class_classification 함수에서 얻은 모델입니다)
    pred_probabilities = iforest_model.decision_function(test_x)

    # ROC 곡선 그리기
    plot_roc_curve(test_y, pred_probabilities)

    # Precision-Recall 곡선 그리기
    plot_precision_recall_curve(test_y, pred_probabilities)
```
